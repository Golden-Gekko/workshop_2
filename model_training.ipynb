## Общие сведения

### Краткое описание проекта

Интернет-магазин собирает историю покупателей, проводит рассылки предложений и
планирует будущие продажи. Для оптимизации процессов надо выделить пользователей,
которые готовы совершить покупку в ближайшее время.

### План работ

- Провести загрузку и предобработку данных (анализ пропусков, приведение типов и т.п.)
- Провести исследовательский анализ данных, построить графики
- Обучить модель
- Получить предсказание на тестовой выборке
- Обобщить результаты исследования

### Описание данных

#### `apparel-messages`

История рекламных рассылок

- `bulk_campaign_id` - идентификатор рекламной кампании (рассылки)
- `client_id` - идентификатор пользователя
- `message_id` - идентификатор сообщений
- `event` - тип действия с сообщением (отправлено, открыто, покупка…)
- `channel` - канал рассылки
- `date` - дата рассылки
- `created_at` - точное время создания сообщения

#### `apparel-purchases`

История покупок

Данные о покупках клиентов по дням и по товарам. В каждой записи покупка
определенного товара, его цена, количество штук.

В таблице есть списки идентификаторов, к каким категориям относится товар. Часто
это вложенные категории (например автотовары-аксессуары-освежители), но также может
включать в начале списка маркер распродажи или маркер женщинам/мужчинам.

Нумерация категорий сквозная для всех уровней, то есть 44 на второй позиции списка
или на третьей – это одна и та же категория. Иногда дерево категорий обновляется, поэтому
могут меняться вложенности, например ['4', '28', '44', '1594'] или ['4', '44', '1594']. 

- `client_id` - идентификатор пользователя
- `quantity` - количество товаров в заказе
- `price` - цена товара
- `category_ids` - вложенные категории, к которым отнсится товар
- `date` - дата покупки
- `message_id` - идентификатор сообщения из рассылки

#### `apparel-target_binary`

Совершит ли клиент покупку в течение следующих 90 дней

- `client_id` - идентификатор пользователя
- `target` - целевой признак

#### `full_campaign_daily_event`

Агрегация общей базы рассылок по дням и типам событий
- `date` - дата
- `bulk_campaign_id` - идентификатор рассылки
- `count_event` - общее количество каждого события event
- `nunique_event` - количество уникальных client_id в каждом событии

#### `full_campaign_daily_event_channel`

Агрегация по дням с учетом событий и каналов рассылки

- `date` - дата
- `bulk_campaign_id` - идентификатор рассылки
- `count_event_channel` - общее количество каждого события по каналам
- `nunique_event_channel` - количество уникальных client_id по событиям и каналам

## Исследовательский анализ данных

```python
def plot_normalized_bar(df, col, figsize=(10, 6), color='seagreen', title=None, xlabel=None, ylabel='Доля'):
    if col not in df.columns:
        raise ValueError(f'Столбец "{col}" не найден в датафрейме')
    
    counts = df[col].value_counts(normalize=True)
    
    plt.figure(figsize=figsize)
    bars = plt.bar(counts.index.astype(str), counts.values, color=color)
    
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2., height,
            f'{height:.4f}',
            ha='center', va='bottom'
        )
    
    if title is None:
        title = f'Нормированное распределение {col}'
    plt.title(title, fontsize=14)
    
    if xlabel is None:
        xlabel = col
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    
    plt.xticks(rotation=45)
    
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()


def plot_histogram(
    df, col, bins='auto', figsize=(10, 6), color='seagreen', title=None, xlabel=None,
    ylabel='Частота', density=False, show_stats=True
):
    if col not in df.columns:
        raise ValueError(f'Столбец "{col}" не найден в датафрейме')
    data = df[col]
    plt.figure(figsize=figsize)
    
    plt.hist(data, bins=bins, color=color, density=density, edgecolor='black', linewidth=1)
    
    if title is None:
        title = f'Гистограмма {col}'
    plt.title(title, fontsize=14)
    
    if xlabel is None:
        xlabel = col
    plt.xlabel(xlabel, fontsize=12)
    
    ylabel_text = 'Плотность вероятности' if density else ylabel
    plt.ylabel(ylabel_text, fontsize=12)
    
    if show_stats:
        stats_text = (f'Объектов: {len(data):,}\n'
                     f'Среднее: {data.mean():.2f}\n'
                     f'Ст. отклонение: {data.std():.2f}\n'
                     f'Медиана: {data.median():.2f}')
        plt.text(0.95, 0.95, stats_text, 
                transform=plt.gca().transAxes,
                ha='right', va='top',
                bbox=dict(facecolor='white', alpha=0.8))
    
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# Преобразование строки, содержащий список в реальный список чисел-категорий
def safe_convert_to_int_list(category_str):
    try:
        raw_list = eval(category_str)
        result = []
        for item in raw_list:
            try:
                num = int(item) if item is not None else None
                if num is not None:
                    result.append(num)
            except (ValueError, TypeError):
                continue
        return result
    except (ValueError, SyntaxError, TypeError):
        return []

df_purch['category_ids'] = df_purch['category_ids'].apply(safe_convert_to_int_list)
```

## Добавление признаков и формирование датасета

### Признаки из `apparel-purchases`

```python
# "Текущая" дата - максимум в данных
df_purch['date'] = pd.to_datetime(df_purch['date'])
current_date = df_purch['date'].max()

client_purch = df_purch.groupby('client_id').agg(
    total_spent=('price', 'sum'),
    avg_purchase_value=('price', 'mean'),
    purchase_count=('client_id', 'count'),
    first_purchase_date=('date', 'min'),
    last_purchase_date=('date', 'max')
).reset_index()

def calc_purchase_frequency(row):
    if row['last_purchase_date'] == row['first_purchase_date']:
        return row['purchase_count']
    return row['purchase_count'] / ((row['last_purchase_date'] - row['first_purchase_date']).days / 30)

# Частота покупок в периоб активности
client_purch['purchase_frequency'] = client_purch.apply(calc_purchase_frequency, axis=1)

# Активный клиент - покупка за последние 180 дней)
client_purch['is_active'] = (current_date - client_purch['last_purchase_date']).dt.days <= 180

client_purch.drop(columns=['first_purchase_date', 'last_purchase_date'], inplace=True)
```

### Подготовка данных

```python
numeric_features = [
    'total_spent', 'purchase_count', 'purchase_frequency',
    'click', 'open', 'purchase', 'send', 
    'is_active_recipient', 'message_frequency'
]

train_df, temp_df = train_test_split(df, test_size=0.4, random_state=RANDOM_STATE, stratify=df['target'])
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=RANDOM_STATE, stratify=temp_df['target'])

preprocessor = ColumnTransformer(
    transformers=[
        ('num',
         Pipeline(
             steps=[
                 ('imputer', SimpleImputer(strategy='median')),
                 ('scaler', StandardScaler())
             ]
         ),
         numeric_features)
    ],
    remainder='passthrough'
)
train_df = pd.DataFrame(
    preprocessor.fit_transform(train_df),
    columns=numeric_features + [col for col in df.columns if col not in numeric_features],
)
val_df = pd.DataFrame(
    preprocessor.fit_transform(val_df),
    columns=numeric_features + [col for col in df.columns if col not in numeric_features],
)
test_df = pd.DataFrame(
    preprocessor.fit_transform(test_df),
    columns=numeric_features + [col for col in df.columns if col not in numeric_features],
)
```

```python
class PandasDataset(Dataset):
    def __init__(self, dataframe):
        self.category_sequences = [
            torch.tensor(seq, dtype=torch.long) 
            for seq in dataframe['category_ids_code']
        ]
        
        self.numeric_features = torch.stack([
            torch.tensor(row.drop(['client_id', 'category_ids_code', 'target']), dtype=torch.float)
            for _, row in dataframe.iterrows()
        ])

        self.targets = torch.tensor(dataframe['target'], dtype=torch.float)
    
    def __len__(self):
        return len(self.numeric_features)
    
    def __getitem__(self, idx):
        return (
            self.numeric_features[idx],
            self.category_sequences[idx],
            self.targets[idx]
        )

    @staticmethod
    def collate_fn(batch):
        numeric_features = torch.stack([item[0] for item in batch])
        category_sequences = [item[1] for item in batch]
        targets = torch.stack([item[2] for item in batch])
        return numeric_features, category_sequences, targets
```

### Модель

```python
class CategoryTransformer(nn.Module):
    def __init__(self, vocab_size, embed_dim=64, num_heads=4, num_layers=2):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim, 
            nhead=num_heads,
            batch_first=True
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
    def forward(self, x, device):
        x = [seq.to(device) for seq in x]
        x_padded = nn.utils.rnn.pad_sequence(x, batch_first=True, padding_value=0)
        mask = (x_padded == 0)
        embeddings = self.embedding(x_padded)
        transformed = self.transformer(embeddings, src_key_padding_mask=mask)
        lengths = torch.tensor([len(seq) for seq in x], device=x_padded.device).float()
        pooled = transformed.sum(dim=1) / lengths.unsqueeze(1)
        return pooled


class TargetPredictor(nn.Module):
    def __init__(self, num_numeric_features, category_vocab_size, embed_dim=64):
        super().__init__()
        self.category_processor = CategoryTransformer(vocab_size=category_vocab_size, embed_dim=embed_dim)

        self.numeric_processor = nn.Sequential(
            nn.Linear(num_numeric_features, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.ReLU()
        )
        
        self.classifier = nn.Sequential(
            nn.Linear(64 + embed_dim, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 1),
            # nn.Sigmoid()
        )
    
    def forward(self, numeric_features, category_ids):
        numeric_processed = self.numeric_processor(numeric_features)
        category_processed = self.category_processor(category_ids, next(self.parameters()).device)
        combined = torch.cat([numeric_processed, category_processed], dim=1)
        prediction = self.classifier(combined)
        return prediction.squeeze()

    @staticmethod
    def collate_fn(batch):
        numeric_features = torch.stack([item[0] for item in batch])
        category_sequences = [item[1] for item in batch]
        targets = torch.stack([item[2] for item in batch])
        return numeric_features, category_sequences, targets
```

### Обучение

```python
def train_model(
    model, train_df, val_df, n_epochs=50, batch_size=32, learning_rate=0.001,
    device='cuda' if torch.cuda.is_available() else 'cpu'
):
    train_loader = DataLoader(
        PandasDataset(train_df),
        batch_size=batch_size,
        shuffle=True,
        collate_fn=PandasDataset.collate_fn
    )
    
    val_loader = DataLoader(
        PandasDataset(val_df),
        batch_size=batch_size,
        shuffle=False,
        collate_fn=PandasDataset.collate_fn
    )
    
    model = model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-4)
    # optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    
    criterion = nn.BCEWithLogitsLoss(torch.tensor([8.0], device=device))
    # criterion = nn.BCEWithLogitsLoss()
    # criterion = nn.BCELoss()

    history = {
        'train_loss': [],
        'val_loss': [],
        'train_auc': [],
        'val_auc': []
    }
    
    for epoch in range(n_epochs):
        # Обучение
        model.train()
        train_loss = 0
        train_preds = []
        train_targets = []

        for numeric, categories, targets in tqdm(train_loader, desc=f'Epoch {epoch+1}/{n_epochs}'):
            numeric = numeric.to(device)
            targets = targets.to(device)
            
            optimizer.zero_grad()
            
            outputs = model(numeric, categories)
            loss = criterion(outputs, targets)
            
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            train_preds.extend(outputs.detach().cpu().numpy())
            train_targets.extend(targets.cpu().numpy())
        
        # Валидация
        model.eval()
        val_loss = 0
        val_preds = []
        val_targets = []
        
        with torch.no_grad():
            for numeric, categories, targets in val_loader:
                numeric = numeric.to(device)
                targets = targets.to(device)
                
                outputs = model(numeric, categories)
                loss = criterion(outputs, targets)
                
                val_loss += loss.item()
                val_preds.extend(outputs.cpu().numpy())
                val_targets.extend(targets.cpu().numpy())
        
        # Метрики
        train_loss /= len(train_loader)
        val_loss /= len(val_loader)
        
        train_auc = roc_auc_score(train_targets, train_preds)
        val_auc = roc_auc_score(val_targets, val_preds)
        
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_auc'].append(train_auc)
        history['val_auc'].append(val_auc)
        
        print(f'Epoch {epoch+1}:')
        print(f'Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}')
        print(f'Train AUC : {train_auc:.4f}  | Val AUC: {val_auc:.4f}')
        print('-' * 50)
        
    # Визуализация обучения
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.title('Loss')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history['train_auc'], label='Train AUC')
    plt.plot(history['val_auc'], label='Val AUC')
    plt.title('ROC-AUC')
    plt.legend()
    
    plt.show()
    
    return model, history
```

```python
def create_metrics_data(y_true, y_predict, start, stop, bins):
    metrics = pd.DataFrame(columns=['Precision', 'Recall', 'Acc', 'F1', 'FP', 'FN', 'Предсказано 1', 'Всего 1'])

    for i in np.linspace(start, stop, num=bins, endpoint=False):
        func = np.vectorize(lambda x: 1 if x > i else 0)
        y_proba = func(y_predict)
        metrics.loc['th_' + str(round(i,2))] = [
            precision_score(y_true, y_proba, zero_division=0),
            recall_score(y_true, y_proba, zero_division=0),
            accuracy_score(y_true, y_proba),
            f1_score(y_true, y_proba),
            confusion_matrix(y_true, y_proba)[0, 1],
            confusion_matrix(y_true, y_proba)[1, 0],
            y_proba.sum(),
            y_true.sum()
        ]
    return metrics

def draw_metrics_data(metrics, title):
    plt.figure(figsize=(15, 7))

    for col in ['Precision', 'Recall', 'Acc', 'F1']:
        plt.plot(metrics[col], label=col)

    plt.grid()
    plt.title(label=title)
    plt.xlabel(xlabel='Порог классификации')
    plt.ylabel(ylabel='Значение')
    plt.legend()
    plt.show()
```
